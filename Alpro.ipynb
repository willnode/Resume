## Pertemuan IV

#### Branching (Kondisi)

```python
# Format Branching di python:
if _condition_:
    statement()
elif _condition_:
    statement()
else:
    statement()
```

#### Evaluasi Truthy

`_condition_` akan mengeksekusi `statement()` dibawahnya jika hasilnya ialah Truthy:

+ Jika `Boolean`, valuenya `True`
+ Jika `Integer`/`Float`, valuenya selain `0`
+ Jika `String`, valuenya selain `""`
+ Jika Object lain, maka valuenya selain `None`

```python
x = 1
if x > 0:
    print("OK") # masuk
if x:
    print("OK") # masuk
if x-x:
    print("NO") # tidak masuk
```

#### If vs Elif

Dua kondisi `if` dapat dicegah agar hanya satu eksekusi yang berjalan menggunakan `elif`: 

```python
x = 100
if x > 100:
    print("A") # masuk
if x > 50:
    print("B") # masuk
if x > 20:
    print("C") # masuk
```

```python
x = 100
if x > 100:
    print("A") # masuk
elif x > 50:
    print("B") # tidak masuk
elif x > 20:
    print("C") # tidak masuk
```


#### Nested Branching

Dua kondisi if dapat dimasukkan satu sama lain:

```python
x = 4
if x % 2 == 0:
    if x > 2:
        print("OK") # masuk
```

... juga dapat digabung jadi satu menggunakan `and`:

```python
x = 4
if x % 2 == 0 and x > 2:
        print("OK") # masuk
```

#### Merged Branching

Dua kondisi if dengan eksekusi yang sama,

```python
x = 4
if x % 2 == 1:
    print("OK") # tidak masuk
elif x > 2:
    print("OK") # masuk
```

... dapat digabung menjadi satu dengan `or`:

```python
x = 4
if x % 2 == 1 or x > 2:
    print("OK") # masuk
```

#### For Loop: 
```python
# format For loop di python
for _index_ in _array_:
    _ekspresi_
```

```python
for i in range(5):
    print (i) # 0, 1, 2, 3, 4
```

#### Argumen range()

Range adalah fungsi iterator (_lihat pertemuan VIII_), dengan argumen:

+ `range(stop)`: Mulai dari `0` (inklusif) hingga `stop` (ekslusif)
+ `range(start, stop)`: Mulai dari `start` (inklusif) hingga `stop` (ekslusif)
+ `range(start, stop, step)` Mulai dari `start` hingga `stop` dengan meloncati tiap index sebanyak `step`.

#### While Loop:

```python
# format While loop di python
while _condition_:
    _ekspresi_
```

Iterasi while harus mempunyai stopping condition agar bisa berhenti dengan sendirinya

Evaluasi `_condition_` sama dengan `if`: Truthy.

#### Break & Continue:

+ `Break` menghentikan loop
+ `Continue` menyekip loop keselanjutnya

```python
for i in range(8):
    if i==1:
        continue
    if i==5:
        break
    print(i) # 0, 2, 3, 4
```

#### Spread Operator

Spread operator adalah operator spesial dimana array dapat dijabarkan (extend) kedalam array diluarnya. Mereka dapat menggantikan `append()`, `extend()`, `update()`.

#### Local & Global Variabel

Variabel yang didalam fungsi (_local variable_) hanya berlaku didalam fungsi saja, sedangkan variabel diluar fungsi (_global variable_) berlaku selama _lifetime_ selama program itu berjalan.

```python
def hitung(n):
    print(x) # 5 (x dari variabel global)
    print(n) # 2
    x = x+n  # 5+2=7
    print(x) # 7 (x ini adalah variabel lokal sekarang)

def okeoke(x):
    print(x) # 8 (x ini dari argumen)
    x += 1
    print(x) # 9

x = 5
hitung(x-3)
okeoke(8)
print(x) # 5 (fungsi tidak bisa mempengaruhi variabel global)
```

Gunakan keyword `global` jika butuh mengubah variabel global:

```python
x, y, z = 9, 8, 5
def yokganti():
    global x, y # hanya butuh mengubah variabel x dan y
    x -= 3 # 9-3
    y = x + z # 6+5
    
yokganti()
print(x, "dan", y) # "6 dan 11"
```

Ingat bahwa list/set/dictionary itu mutable. Perubahan didalam list (dengan fungsi append/extend/dsb.) akan berefek tidak peduli itu lokal atau global:

```python
def yokisi(r):
    r.extend([*range(3)])
    
arr = [0, 1]
yokisi(arr)
print(arr) # [0, 1, 0, 1, 2]
```

#### Sets (_Himpunan_)

+ Sets adalah array dengan koleksi yang dijamin unik.
+ Sets adalah mutable, namun jika ada 2 elemen yang sama, ia akan ditumpang tindihkan.
+ Urutan sets tidak dijamin, sehingga sets tak bisa diakses lewat index (hanya bisa lewat iterasi dan for loop)

```python
s = {1, 2, 3}    # sets (pakai kurawal)
s[1] = 3         # RUNTIME ERROR
print(s[1])      # RUNTIME ERROR

s = {3, 4, 3, 1} # sets baru
print(s)         # {1, 3, 4}
print(4 in s)    # True
```

Terlihat seperti tak berguna, namun sets sangat membantu dalam memproses dan menjaring data dan himpunan.

#### Operasi Sets

+ Operasi Union (Penggabungan) `|`
  - (dapat digunakan untuk menambah elemen)
+ Operasi Intersect (Irisan) `&`
  - (dapat digunakan untuk mencari persekutuan)
+ Operasi Difference (Selisih) `-`
  - (dapat digunakan untuk membuang elemen)
+ Operasi XOR (Komplemen) `^`
  - (dapat digunakan untuk mencari perbedaan)

```python
a = {1, 2, 3, 6}
b = {2, 3, 4, 6}
c = {1, 2, 7, 8}

print (a | b) # {1, 2, 3, 4, 6}
print (a & b) # {2, 3, 6}
print (a ^ b) # {1, 4}
print (a - c) # {3, 6}
print ((a | c) - b) # {8, 1, 7}
```

#### Dictionary (_Kamus_)

Dictionary adalah list dengan indexer sembarang tipe data, layaknya sets dengan pasangan datanya.

```python
harga = {"Coklat":5000, "Stroberi":2000}
print(harga["Coklat"]) # 5000
```

Dictionary itu Mutable. Menambahkan/Mengganti data Dictionary dapat dilakukan dengan mengganti via key.

```python
harga = {"Coklat":5000, "Stroberi":2000}
harga["Susu"] = 3000
print(harga) # {"Coklat":5000, "Stroberi":2000, "Susu":3000}
harga["Coklat"] = 7000
print(harga) # {"Coklat":7000, "Stroberi":2000, "Susu":3000}
```

Jika dictionary key tak ditemukan, maka runtime error. Gunakan `.get()` untuk mencegahnya.

```python
harga = {"Coklat":5000, "Stroberi":2000}
print(harga["Susu"]) # RUNTIME ERROR
print(harga.get("Susu", 0)) # OK (0)
```

#### Fungsi Dictionary

+ `.clear()`: Menghapus semua elemen Dictionary
+ `.copy()`: Menduplikat Dictionary
+ `.get(k[,d])`: Mendapatkan value dari key `k`
  - Kalau tidak ada, maka return default `d`, tidak error
+ `.items()`: Buat iterator key+value dari Dictionary
